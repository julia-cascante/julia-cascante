Import libraries

```python
import matplotlib.pyplot as plt
import numpy as np
```

I will be setting up my constants

```python
g = 9.81  # acceleration due to gravity in m/s^2
v0 = 0.0  # initial velocity in m/s
y0 = 100.0  # initial height in meters
dt = 0.1  # time step in seconds
total_time = 20.0  # total simulation time in seconds
```

Arrays to store time, velocity, and position

```python
t_values = np.arange(0, total_time, dt)
y_values = np.zeros_like(t_values)
v_values = np.zeros_like(t_values)
```

Initial conditions

```python
y_values[0] = y0
v_values[0] = v0
```

Numerical integration using Euler's method

```python
for i in range(1, len(t_values)):
    v_values[i] = v_values[i-1] - g * dt
    y_values[i] = y_values[i-1] + v_values[i-1] * dt
```

Plotting

```python
plt.figure(figsize=(10, 6))
plt.plot(t_values, y_values, label='Position (m)')
plt.plot(t_values, v_values, label='Velocity (m/s)')
plt.title('Free Fall with Gravity')
plt.xlabel('Time (s)')
plt.ylabel('Height (m) / Velocity (m/s)')
plt.legend()
plt.grid(True)
plt.show()
```
